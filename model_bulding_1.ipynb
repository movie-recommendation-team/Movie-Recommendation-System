```python
# prompt: please remove the na values

movies = movies.dropna()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_features = ["popularity", "budget", "revenue", "runtime"]
movies[numerical_features] = scaler.fit_transform(movies[numerical_features])
```

```python
movies
```

```python
movies['combined_col'] = movies['title'].astype(str) + " " + movies['overview'].astype(str) + " " + movies['original_language'].astype(str)
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
movies.reset_index(drop=True, inplace=True)
```

```python
text_embeddings = model.encode(movies["combined_col"], show_progress_bar=True)
```

```python
movies.drop(['vote_average', 'vote_count'], axis=1, inplace=True)
```

```python
movies
```

```python
text_embeddings = np.array(text_embeddings)
```

```python
movie_embeddings = np.hstack((text_embeddings, movies[numerical_features].values))
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=10, metric="cosine", algorithm="brute")
```

```python
model = knn.fit(movie_embeddings)
```

```python
def get_recommendations(movie_title):
    idx = movies[movies["title"] == movie_title].index[0]  # Find index of the movie
    distances, indices = knn.kneighbors([movie_embeddings[idx]], n_neighbors=10)

    # Retrieve recommended movie titles (excluding the input movie itself)
    recommended_movies = [movies.iloc[i]["title"] for i in indices.flatten()[1:]]
    return recommended_movies
```

```python
print(get_recommendations("Kedarnath"))
```

```python
print(get_recommendations("Interceptor"))
```
