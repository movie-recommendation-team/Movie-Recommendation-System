```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import MinMaxScaler
import pickle
```

```python


parquet_file = '/content/part-00000-bcc51f97-70f3-4315-932d-dca33a0551ca-c000.snappy.parquet'
csv_file = '/content/output.csv'  # Specify the desired output CSV file path

try:
    df = pd.read_parquet(parquet_file)
    df.to_csv(csv_file, index=False)  # Set index=False to avoid writing row indices to the CSV
    print(f"Successfully converted '{parquet_file}' to '{csv_file}'")
except FileNotFoundError:
    print(f"Error: Parquet file '{parquet_file}' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
movies = pd.read_csv('/content/output.csv')
```

```python
movies.dtypes
```

```python
movies
```

```python
movies['genres'] = movies['genres'].fillna("")
movies['credits'] = movies['credits'].fillna("")

movies
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isna().sum()
```

```python
movies['vote_average'] = pd.to_numeric(movies['vote_average'], errors='coerce')
movies['vote_count'] = pd.to_numeric(movies['vote_count'], errors='coerce')
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.dropna(subset=['vote_average'], inplace=True)

movies.isna().sum()
```

```python
generous_m = movies['vote_count'].quantile(0.70)
C = movies['vote_average'].mean()

def weighted_rating(x, m=generous_m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (v + m) * C)

movies['weighted_rating'] = movies.apply(weighted_rating, axis=1)
```

```python
movies.drop(['vote_count', 'vote_average'], axis=1, inplace=True)
```

```python
scaler = MinMaxScaler()
numerical_features = ["popularity", "budget", "revenue", "runtime", 'weighted_rating', "release_year"]

movies[numerical_features] = scaler.fit_transform(movies[numerical_features])
```

```python
movies
```

```python
movies.dropna(subset=['title'], inplace=True)
```

```python
movies['combined_text'] = movies['title'] + " " + movies['overview'] + " " + movies['original_language'] + " " + movies['genres'] + " " + movies['credits']
```

```python
movies
```

```python
movies.isna().sum()
```

```python
# movies.reset_index(drop=True, inplace=True)
```

```python
movies.isna().sum()
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

text_embeddings = model.encode(movies["combined_text"].to_list(), show_progress_bar=True)
```

```python
text_embeddings = np.array(text_embeddings)

movie_embeddings = np.hstack((text_embeddings, movies[numerical_features].values))
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=10, metric="cosine", algorithm="brute")

model = knn.fit(movie_embeddings)
```

```python
def get_recommendations(movie_title):
    idx = movies[movies["title"] == movie_title].index[0]
    distances, indices = knn.kneighbors([movie_embeddings[idx]], n_neighbors=9)

    recommended_movies = [movies.iloc[i]["title"] for i in indices.flatten()[1:]]
    return recommended_movies
```

```python
print(get_recommendations("Kedarnath"))
```

```python
# Save the trained model using pickle
with open('recommendation_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Save the movie embeddings
with open('movie_embeddings.pkl', 'wb') as f:
  pickle.dump(movie_embeddings, f)

# Save the movies DataFrame
with open('movies_df.pkl', 'wb') as f:
    pickle.dump(movies, f)
```
